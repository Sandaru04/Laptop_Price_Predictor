```python
# convert datatype weight and ram into int32 and float32
data_frame['Weight'] =data_frame['Weight'].str.replace('kg', '', case=False).astype('float32')
data_frame['Ram'] =data_frame['Ram'].str.replace('GB', '').astype('int32')
```

```python
def add_company(input):
    if input =='Samsung' or input =='Razer' or input == 'Mediacom' or input == 'Microsoft' or input == 'Vero' or input =='Xiaomi' or input == 'Chuwi' or input == 'Fujitsu' or input == 'Google' or input == 'LG' or input =='Huawei':
        return 'Other'
    else :
        return input

data_frame['Company'] = data_frame['Company'].apply(add_company)
    
```

```python
data_frame['TouchScreen']=data_frame['ScreenResolution'].apply(lambda x:1 if 'TouchScreen' in x else 0)
data_frame['IPS']=data_frame['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
data_frame['cpu_name'] = data_frame['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
data_frame['cpu_name'].value_counts()
```

```python
def set_processor(name):
    if name == 'Intel Core i7' or name == 'Intel Core i5' or name == 'Intel Core i3':
        return name
    else:
        if name.split()[0] == 'AMD':
            return 'AMD'
        else:
            return 'Other'
```

```python

data_frame['cpu_name'] = data_frame['cpu_name'].apply(set_processor)
```

```python
data_frame['gpu_name'] = data_frame['Gpu'].apply(lambda x:" ".join(x.split()[0:1]))
```

```python
def set_os(inpt):
    if inpt == 'Windows 10' or inpt == 'Windows 7' or inpt == 'Windows 10 S':
        return 'Windows'
    elif inpt == 'macOS' or inpt == 'Mac OS X':
        return 'Mac'
    elif inpt == 'Linux':
        return inpt
    else:
        return 'Other'

data_frame['OpSys'] = data_frame['OpSys'].apply(set_os)

data_frame['OpSys'].value_counts()
```
